```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
def fx(u):
    return np.cos(u**2)**2 + np.pi

xx = np.linspace(-2.1,2.1,211)

plt.plot(xx,fx(xx));
plt.grid();
plt.show();
```

```python
#target
a = 1

x0 = np.array([a-1,a+1])

# iterations
iterations = 10
limitVals = np.zeros((iterations,2))
xAxisVals = np.zeros((iterations,2))

# zeno's paradox algorithm
for i in range(iterations):
    
    xAxisVals[i,:] = x0
    limitVals[i,:] = fx(x0)
    
    x0 = (x0+a)/2
```

```python
print('Limit from the left')
print(np.vstack((xAxisVals[:,0],limitVals[:,0])).T)

print('Limit from the right')
print(np.vstack((xAxisVals[:,1],limitVals[:,1])).T)

print(f'value of f(x) at x={a}:')
print(fx(a))
```

```python
plt.plot(xx,fx(xx),'k')
plt.plot([a,a],[np.pi,1+np.pi],'k--',linewidth=.2)
plt.plot(xAxisVals,limitVals,'o',markerfacecolor='w')
plt.title(f'f({a}) = {fx(a)}')
plt.xlim([a-1,a+1])
plt.show()
```
