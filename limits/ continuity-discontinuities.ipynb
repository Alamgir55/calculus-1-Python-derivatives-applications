```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
def fx(m):
    m = np.array(m)
    y = np.zeros(m.shape)
    
    tol = 10e-4
    
    y[m<tol] = np.sin(m[m<tol]*np.pi)
    y[m>tol] = -(m[m>tol]-2)**2
    y[np.abs(m)<tol] = 1.5
    return y

xx = np.linspace(-1,2,10001)
plt.plot(xx,fx(xx),'o')
plt.show()
```

```python
xFromLeft = -np.logspace(np.log10(.1),np.log10(.002),10)
xFromRight = np.logspace(np.log10(.002),np.log10(.1),10)

limitFromLeft = fx(xFromLeft)
limitFromRight = fx(xFromRight)

print(f'Limit as x->0 from the left is {limitFromLeft[-1]}')
print(f'Limit as x->0 from the right is {limitFromRight[0]}')
print(f'Value of the fucntion at x=0 is {fx(0)}')

plt.plot(xx,fx(xx),'k.')
plt.plot(xFromLeft,limitFromLeft,'s')
plt.plot(xFromRight,limitFromRight,'o')
plt.plot(0,fx(0),'rx')
plt.xlim([-.1,.1])
plt.show()
```

```python
from sympy.abc import x

piece1 = sym.sin(x*sym.pi)
piece2 = 1.5
piece3 = -(x-2)**2

fx = sym.Piecewise(
    (piece1,x<0),
    (piece2,sym.Eq(x,0)),
    (piece3,x>0)
)
fx.subs(x,0)
fx
```

```python
print('Limit as x->0 from the left:')
print(sym.limit(fx,x,0,dir='-'))


print('Limit as x->0 from the right:')
print(sym.limit(fx,x,0,dir='+'))
```

```python
fx = 3 / (1-x**2)

sym.plot(fx,(x,-2,2),ylim=[-10,10])
```

```python
print('Limit as x->-1 from the left:')
print(sym.limit(fx,x,-1,dir='-'))


print('Limit as x->-1 from the right:')
print(sym.limit(fx,x,-1,dir='+'))

print('Two-sided limit as x->-1:')
print(sym.limit(fx,x,-1,dir='+'))
```
