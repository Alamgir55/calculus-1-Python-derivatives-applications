```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
def fx(u):
    return np.cos(u*np.pi) + np.log(u)**2

xx = np.linspace(.1,2*np.pi,213)

plt.plot(xx,fx(xx))
plt.grid()
plt.show()
```

```python
expFact = 4
g = 1-np.linspace(1,.00001**expFact,10)**(1/expFact)

xFromLeft  = 3-g
xFromRight = 3+g

limitFromLeft  = fx(xFromLeft)
limitFromRight = fx(xFromRight)

print(f'Limit approcah {limitFromLeft[0]} from the left.')
print(f'Limit approcah {limitFromRight[0]} from the left.')
print(f'Function value: {fx(3)} at x=3')

plt.plot(xx,fx(xx),label='f(x)')
plt.plot(xFromLeft,limitFromLeft,'s',label='From the left')
plt.plot(xFromRight,limitFromRight,'s',label='From the left')
plt.legend()
plt.show()
```

```python
from sympy.abc import x

sfx = sym.cos(x*sym.pi) + sym.log(x)**2

print(sfx.subs(x,3))
print(fx(3))
```

```python
sym.limit(sfx,x,3,dir='+-')
```

```python
sfx2 = 1/(x-2)**2

sfx2_lam = sym.lambdify(x,sfx2)

xx = np.linspace(0,4,1000)
plt.plot(xx,sfx2_lam(xx))
plt.ylim([0,100])
plt.plot([2,2],[0,100],'--',color=[.6,.6,.6])
plt.show()
```

```python
print('Limit as x->2 from the left:  ' + str(sym.limit(sfx2,x,2,dir='-')))
print('Limit as x->2 from the right:  ' + str(sym.limit(sfx2,x,2,dir='+')))
print('Limit as x->2 (two-sided):  ' + str(sym.limit(sfx2,x,2,dir='+-')))
```
