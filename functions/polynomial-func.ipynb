```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
coefs = np.random.randn(4)
coefs

xx = np.linspace(-5,5,101)
y = np.zeros(len(xx)) 

fname = '$y = '
for i, c in enumerate(coefs):
    y += c*xx**i
    fname += '+ '[int(c<0)] + f'{c:.2f}x^{i}'
    
plt.plot(xx,y)
plt.title(fname + '$')
plt.show()
```

```python
from sympy.abc import x

top = x**2 - 2*x
bot = x**2 - 4
sy = top / bot

sym.plot(sy,(x,-3,3),ylim=[-10,10],title=f'$y = {sym.latex(sy)}$')
```
