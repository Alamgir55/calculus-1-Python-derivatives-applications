```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

from IPython.display import display,Math
```

```python
n = [1,2,5,10]

for i in n:
    e = (1+(1/i))**i
    
    print(f'n: {i:2.0f}, est.e: {e:.5f}, diff to e: {np.exp(1)-e:.5f}')
```

```python
n = np.arange(1,1001)

eDiffs = np.exp(1) - (1+1/n)**n

plt.plot(n,eDiffs)

```

```python
xDomain = [-2, 2]

x = np.linspace(xDomain[0],xDomain[1],53)

y1 = np.exp(x)
y2 = np.exp(x**2)
y3 = np.exp((-x)**2)
y4 = np.exp(-(x**2))
y5 = np.exp(x)**2

plt.figure(figsize=(8,6))
plt.plot(x, np.exp(x), label='$y = e^x$')
plt.ylim(-1, 11)
plt.plot(x, np.exp(x**2), label='$y = e^{x^2}$')
plt.plot(x, np.exp((-x)**2), label='$y = e^{(-x)^2}$')
plt.plot(x, np.exp(-(x**2)), label='$y = e^{-(x^2)}$')
plt.plot(x, np.exp(x)**2, label='$y = (e^x)^2$')
plt.xlabel('x')
plt.grid()
plt.ylabel('y = f(x)')
plt.legend()
plt.show()
```

```python
s_beta = sym.var('beta')

s_y = sym.exp(s_beta) - sym.log(s_beta) - np.exp(1)

sym.plot(s_y,(s_beta,xDomain[0],xDomain[1]),
         title=f'$f(\\beta) = {sym.latex(s_y)}$',
         ylabel='$y=f(\\beta)$',xlabel='$\\beta$');
```

```python
x = np.linspace(-4,4,30)
plt.plot(x,np.exp(x),linewidth=2,label='exp(x)')

plt.plot(x,np.log(np.exp(x)),linewidth=2,label='log(exp(x))')
plt.plot(x,np.exp(np.log(x)),'o', linewidth=2,label='exp(log(x))')

x = np.linspace(.001,4,30)
plt.plot(x,np.log(x),linewidth=2,label='log(x)')

plt.legend()
plt.axis('square')
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
