```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import sympy as sym
```

```python
xDomain = [-2, 2]

resolution = .1
x = np.arange(xDomain[0], xDomain[1]+resolution,resolution)

x = np.linspace(xDomain[0], xDomain[1],41)

y = x**2 + 3*x**3 - x**4

plt.plot(x,y,label='$y=x^2+3x^3-x^4$')
plt.grid()
plt.legend()
plt.xlim(xDomain[0], xDomain[1])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
s_beta = sym.var('beta')

s_y = s_beta**2 + 3*s_beta**3 - s_beta**4

sym.plot(s_y,(s_beta,xDomain[0],xDomain[1]),xlabel='x',ylabel=None,
         title=f'$f(\\beta) = {sym.latex(s_y)}$')
```

```python
fx = sym.lambdify(s_beta,s_y)

yy = fx(x)

plt.plot(x,yy,label='$y=x^2+x^3-x^4$')
plt.grid()
plt.legend()
plt.xlim(xDomain[0], xDomain[1])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
