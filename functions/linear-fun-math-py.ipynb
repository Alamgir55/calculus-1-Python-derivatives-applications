```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
xDomain = [-2, 2]

resolution = .1
x = np.arange(xDomain[0], xDomain[1]+resolution,resolution)

x = np.linspace(xDomain[0], xDomain[1],41)

y = x**2 + 3*x**3 - x**4

plt.plot(x,y,label='$y=x^2+3x^3-x^4$')
plt.grid()
plt.legend()
plt.xlim(xDomain[0], xDomain[1])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
