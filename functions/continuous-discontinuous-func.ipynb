```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
resolution = .1
xx = np.arange(-1,2,resolution)

y =  [0]*3
y[0] = np.sin(xx*np.pi)
y[1] = 1.5*np.ones(len(xx))
y[2] = -(xx-2)**2

xdomains = [0]*3
xdomains[0] = xx<0
xdomains[1] = np.abs(xx)<resolution/2
xdomains[2] = xx>0

marker = '-o-'
for i in range(len(y)):
    plt.plot(xx[xdomains[i]],y[i][xdomains[i]],marker[i])

plt.show()
```

```python
from sympy.abc import x

fx = sym.Piecewise(
    (sym.sin(x*sym.pi),x<0),
    (1.5,sym.Eq(x,0)),
    (-(x-2)**2,x>0)
)
sym.plot(fx,(x,xx[0],xx[-1]))
```

```python
fx.subs(x,0.00001)
```

```python
y = np.sin(xx*np.pi) + xx**2
y[np.argmin(np.abs(xx))] = np.pi

plt.plot(xx,y,'o')
```
