```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(-3,5,100)

piece1 = 0        * (xx<0)
piece2 = -2*xx    * ((xx>=0) & (xx<3))
piece3 = xx**3/10 * (xx>3)

y = piece1+piece2+piece3
plt.plot(xx,y)
```

```python
plt.plot(xx[xx<0],np.zeros(np.sum(xx<0)),label='Picee 1')
plt.plot(xx[(xx>=0) & (xx<3)],-2*xx[(xx>=0) & (xx<3)],label='Piece 2')
plt.plot(xx[xx>=3],xx[xx>=3]**3/10,label='Piece 3')

plt.show()
```

```python
from sympy.abc import x

piece1 = 0 
piece2 = -2*x
piece3 = x**3/10

fx = sym.Piecewise((piece1,x<0),(piece2,(x>=0)&(x<3)),(piece3,x>=3))

sym.plot(fx,(x,xx[0],xx[-1]),title='Piecewise function',
             xlabel='x',ylabel='y')
```

```python
xloc = .5

fx.subs(x,xloc)
```
