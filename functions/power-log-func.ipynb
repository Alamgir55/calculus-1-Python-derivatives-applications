```python
a = np.random.rand(100)
b = np.random.rand(100)

# np.log(a*b) = np.log(a)+np.log(b)
v = np.log(a*b) - (np.log(a)+np.log(b))

plt.plot(v,'s-')
```

```python
a = np.random.rand(100)
b = np.random.rand(100)

# np.log(a*b) = np.log(a)+np.log(b)
v = np.log(a**b) - a*np.log(b)

plt.plot(v,'s-')
```

```python
a = np.random.rand(100)
b = np.random.rand(100)

# np.log(a*b) = np.log(a)+np.log(b)
v = np.log(a/b) - (np.log(a) - np.log(b))

plt.plot(v,'s-')
```
