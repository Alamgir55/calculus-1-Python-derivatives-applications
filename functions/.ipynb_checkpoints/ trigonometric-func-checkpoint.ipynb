```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(-np.pi,2*np.pi,1234)

f = [0]*3

f[0] = np.sin(xx) , 'sin(x)'
f[1] = np.sin(xx)**2 , 'sin(x)**2'
f[2] = np.sin(xx**2), 'sin(x**2)'

for y,label in f:
    plt.plot(xx,y,label=label)

plt.legend()
plt.xlabel('Angle (red.)')
plt.ylabel('y = f(x)')
plt.show()
```

```python
xx = np.linspace(-np.pi,2*np.pi,1234)

f = [0]*3

f[0] = np.sin(np.cos(xx)) , 'sin(cos(x))'
f[1] = np.cos(np.sin(xx)) , 'cos(sin(x))'
f[2] = np.cos(xx)         , 'cos(x)'

for y,label in f:
    plt.plot(xx,y,label=label)

plt.legend()
plt.xlabel('Angle (red.)')
plt.ylabel('y = f(x)')
plt.show()
```

```python
for i in range(2,5):
    xx = np.linspace(0,2*np.pi,10**i)
    plt.plot(xx,np.tan(xx),'.-',label=f'{10**i} pnts')

plt.xlim([1.55,1.6])
```
